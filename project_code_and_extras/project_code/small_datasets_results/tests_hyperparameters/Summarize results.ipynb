```python
import os
import pandas as pd
```

```python
files_list = os.listdir('saved_csv/')
files_list = ['saved_csv/'+file for file in files_list]
files_list
```

```python
df_list = []

for file in files_list:
    
    temp_df = pd.read_csv(file)
    if file.startswith('saved_csv/mnist'):
        temp_df['Dataset'] = 'MNIST'
    else: 
        temp_df['Dataset'] = 'air pollution'
    df_list.append(temp_df)
    
df = pd.concat(df_list, ignore_index=True)
df = df[['Dataset', 'Model name', 'variants_name', 'final_validation_loss']]
```

```python
df['variants_name'] = df['variants_name'].apply(lambda x: x.replace('-','_'))
```

```python
available_models = ['classic_AE_','GCN_AE_','GCN_AE_GCN_']

df['Model name'] = df['variants_name'].apply(lambda x: available_models[0] if x.startswith(available_models[0])
                                                  else available_models[2] if x.startswith(available_models[2])
                                                  else available_models[1] if x.startswith(available_models[1])
                                                  else None )
```

```python
df['variants_name'] = df.apply(lambda x: x['variants_name'].replace(x['Model name'],''), axis=1)
df = df.rename(columns={'variants_name':'parameter_value'})
```

```python
tuning_parameter = ['latDim','EmbSeq','batchSize']

df['Tuning_parameter'] = df['parameter_value'].apply(lambda x: tuning_parameter[0] if x.startswith(tuning_parameter[0])
                                                  else tuning_parameter[2] if x.startswith(tuning_parameter[2])
                                                  else tuning_parameter[1] if x.startswith(tuning_parameter[1])
                                                  else 'EmbSeq' )
```

```python
df['Tuning_parameter'] = df['Tuning_parameter'].apply(lambda x: x[0].upper()+x[1:])
df['Model name'] = df['Model name'].apply(lambda x: x[:-1])
df['parameter_value'] = df['parameter_value'].apply(lambda x: '-' if x=='Sequential' else x)
df['final_validation_loss'] = df['final_validation_loss'].apply(lambda x: round(x,4))
```

```python
df = df[['Tuning_parameter', 'Dataset', 'Model name', 'parameter_value', 'final_validation_loss']]
```

```python
df = df.sort_values(by=['Tuning_parameter','Dataset','final_validation_loss'], ascending=[False, False, True], ignore_index=True)
```

```python
df.to_csv('Tuning_parameters.csv',index=False)
```
