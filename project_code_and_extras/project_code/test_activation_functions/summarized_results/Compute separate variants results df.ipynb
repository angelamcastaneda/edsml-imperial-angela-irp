```python
import pandas as pd
```

```python
pairs_tests_list = [['1',''],['2',''],['3','1'],['3','2'],['4','1'],['4','2']]
```

### Create two dataframes, with variants results ranked for each dataset

```python
mnist_files_locations = ['../model'+model_number+'/saved_csv/model'+model_number+
                         '_mnist_test'+test_number+'_variants_losses.csv' 
                         for model_number, test_number in pairs_tests_list ] 
```

```python
mnist_list_dfs = []

for i in range(len(mnist_files_locations)):
    df = pd.read_csv(mnist_files_locations[i])
    mnist_list_dfs.append(df)
    
mnist_results = pd.concat(mnist_list_dfs, ignore_index=True)

mnist_results.head()
```

```python
air_pollution_files_locations = ['../model'+model_number+'/saved_csv/model'+model_number+
                         '_air_pollution_test'+test_number+'_variants_losses.csv' 
                         for model_number, test_number in pairs_tests_list ] 
```

```python
air_pollution_list_dfs = []

for i in range(len(air_pollution_files_locations)):
    df = pd.read_csv(air_pollution_files_locations[i])
    air_pollution_list_dfs.append(df)
    
air_pollution_results = pd.concat(air_pollution_list_dfs, ignore_index=True)

air_pollution_results.head()
```

### Read the variants details and paste them into the previous tables

```python
variants_details = pd.read_csv('variants_details.csv')

variants_details.head()
```

```python
air_pollution_results = pd.merge(left=air_pollution_results,right=variants_details,left_on=['variants_name','Model name'],right_on=['Variant name','Model number'],how='left')

air_pollution_results.head()
```

```python
air_pollution_results = air_pollution_results[['Model name','Sequence of activation functions','final_validation_loss']]

air_pollution_results['final_validation_loss'] = air_pollution_results['final_validation_loss'].round(4)
air_pollution_results['ranking'] = air_pollution_results['final_validation_loss'].rank()

air_pollution_results.to_csv('air_pollution_results.csv')
air_pollution_results
```

### And do the same for MNIST

```python
mnist_results = pd.merge(left=mnist_results,right=variants_details,left_on=['variants_name','Model name'],right_on=['Variant name','Model number'],how='left')

mnist_results = mnist_results[['Model name','Sequence of activation functions','final_validation_loss']]
mnist_results['final_validation_loss'] = mnist_results['final_validation_loss'].round(4)
mnist_results['ranking'] = mnist_results['final_validation_loss'].rank()

mnist_results.to_csv('mnist_results.csv')
mnist_results
```
