```python
import pandas as pd
```

```python
pairs_tests_list = [['1',''],['2',''],['3','1'],['3','2'],['4','1'],['4','2']]
```

### Create two dataframes, with variants results ranked for each dataset

```python
mnist_files_locations = ['../model'+model_number+'/saved_csv/model'+model_number+
                         '_mnist_test'+test_number+'_variants_losses.csv' 
                         for model_number, test_number in pairs_tests_list ] 
```

```python
mnist_list_dfs = []

for i in range(len(mnist_files_locations)):
    df = pd.read_csv(mnist_files_locations[i])
    mnist_list_dfs.append(df)
    
mnist_results = pd.concat(mnist_list_dfs, ignore_index=True)
```

```python
air_pollution_files_locations = ['../model'+model_number+'/saved_csv/model'+model_number+
                         '_air_pollution_test'+test_number+'_variants_losses.csv' 
                         for model_number, test_number in pairs_tests_list ] 
```

```python
air_pollution_list_dfs = []

for i in range(len(air_pollution_files_locations)):
    df = pd.read_csv(air_pollution_files_locations[i])
    air_pollution_list_dfs.append(df)
    
air_pollution_results = pd.concat(air_pollution_list_dfs, ignore_index=True)
```

### Join the two dataframes right next to each other

```python
mnist_results = mnist_results.rename(columns={'variants_name':'MNIST_variants_ranked',
                                              'final_validation_loss':'MNIST_validation_loss'})
```

```python
air_pollution_results = air_pollution_results.rename(columns={'variants_name':'air_pollution_variants_ranked',
                                              'final_validation_loss':'air_pollution_validation_loss'})
```

```python
air_pollution_results = air_pollution_results.drop('Model name', axis=1)
```

```python
df_complete = pd.concat([mnist_results, air_pollution_results], axis=1)
df_complete
```

```python
df_complete.to_csv('variants_results.csv', index=False )
```
