# Example using the Y axis slice

```python
import pyvista as pv
from math import floor
```

```python
[i for i in range(40,100,8)]
```

```python
#NO Template
# template = pv.read(f'../../Datasets/WaveSuite_VTK/sateNo4_1_40.vtu')

pl = pv.Plotter(shape=(2, 4))

for j, num in enumerate(range(40,100,8)): 
    pl.subplot(floor(j/4), j%4)
    
    mesh0 = pv.read(f'../../Datasets/WaveSuite_VTK/sateNo4_1_{num}.vtu')
    
    mesh = mesh0.copy()
    mesh['nut'] = mesh0['nut']
    
    mesh.set_active_scalars('nut')    
    single_slice = mesh.slice(normal=[0, 1, 0])
    pl.add_mesh(single_slice, clim=[0, 1.5] , cmap='viridis') #cmap=cmap, , show_edges=True
    pl.camera_position = 'xz'
    pl.camera.zoom(1.7)
    pl.add_text(f'{j}_{num}')
pl.show()
```

### So what if there is a problem because there are two NUT variables in the files? 
Then, I will rename the 'nut' cell variable to 'nut_cell' to make sure that later when I call 'nut I get the point variable

```python
[i for i in range(40,100,8)]
```

```python
template = pv.read(f'../../Datasets/WaveSuite_VTK/sateNo4_1_40.vtu')
template.rename_array(old_name='nut', new_name='nut_cell', preference='cell')

pl = pv.Plotter(shape=(2, 4))

for j, num in enumerate(range(40,48)): #range(40,100,8)
    pl.subplot(floor(j/4), j%4)
    
    mesh0 = pv.read(f'../../Datasets/WaveSuite_VTK/sateNo4_1_{num}.vtu')
    mesh0.rename_array(old_name='nut', new_name='nut_cell', preference='cell')
    
    mesh = template.copy()
    mesh['nut'] = mesh0['nut']
    
    mesh.set_active_scalars('nut')    
    single_slice = mesh.slice(normal=[0, 1, 0])
    pl.add_mesh(single_slice, clim=[0, 1.5] , cmap='viridis') #cmap=cmap, , show_edges=True
    pl.camera_position = 'xz'
    pl.camera.zoom(1.7)
    pl.add_text(f'{j}_{num}')
pl.show()
```
